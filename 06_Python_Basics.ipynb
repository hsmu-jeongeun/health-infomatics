## Part 1. 변수와 기본 연산

```python
# 변수에 값 저장하기
patient_name   = '홍길동'
patient_age    = 52
body_temp      = 37.2  # 섭씨
heart_rate     = 88    # bpm

print(f'환자명: {patient_name}')
print(f'나이: {patient_age}세')
print(f'체온: {body_temp}°C')
print(f'심박수: {heart_rate} bpm')

# 기본 사칙연산
print('\n=== 기본 연산 ===')
print(f'체온(°C → °F 변환): {body_temp * 9/5 + 32:.1f}°F')
print(f'BMI 계산 (체중 70kg, 신장 1.72m): {70 / (1.72**2):.1f}')
```

```python
# 변수의 데이터 타입 확인하기
print(f"{type(patient_name)}, {type(patient_age)}, {type(body_temp) }, {type(heart_rate) }")
```

## Part 2. 리스트(List) — 여러 데이터를 한 곳에

```python
# 1주일치 체온 기록 (리스트)
temp_week = [36.8, 37.1, 38.3, 38.7, 37.9, 37.2, 36.9]
days      = ['월', '화', '수', '목', '금', '토', '일']

print('1주일 체온 기록:')
for i, (day, temp) in enumerate(zip(days, temp_week)):
    status = '발열' if temp >= 37.5 else '정상'
    print(f'  {day}요일: {temp}°C  [{status}]')

# 기본 통계
print(f'\n평균 체온: {sum(temp_week)/len(temp_week):.2f}°C')
print(f'최고 체온: {max(temp_week)}°C')
print(f'최저 체온: {min(temp_week)}°C')
```

## Part 3. 조건문(if) — 체온에 따른 처치 결정

```python
# =====================
# 수정해 보세요!
# =====================
current_temp = 38.7  # 현재 환자 체온 (값을 바꿔보세요)
# =====================

# 체온에 따른 간호 중재 결정 로직
if current_temp >= 39.0:
    action = '즉시 의사 보고 + 해열제 투약 + 냉각 처치'
elif current_temp >= 37.5:
    action = '30분 후 재측정 + 수분 섭취 격려 + 의사 보고'
elif current_temp >= 36.0:
    action = '정상 범위 — 계속 모니터링'
else:
    action = '저체온 — 보온 처치 + 즉시 보고'

print(f'현재 체온: {current_temp}°C')
print(f'권장 중재: {action}')
```

## Part 4. 반복문(for) — 30일치 혈당 데이터 분석

> 참고 : 30명 환자가 아니라, 한 환자의 30일 혈당 데이터

```python
# 30일치 혈당 데이터 (mg/dL)
# 공복혈당 정상: 100 미만 / 당뇨 전단계: 100-125 / 당뇨: 126 이상
blood_glucose_30days = [
    95, 102, 118, 131, 125, 98, 143, 156, 112, 99,
   138, 121, 144, 167, 109, 97, 128, 135, 148, 103,
   122, 141, 158, 116, 94, 133, 119, 151, 107, 126
]

print(f'총 측정 일수: {len(blood_glucose_30days)}일')
print(f'혈당 데이터: {blood_glucose_30days}')
```

```python
# =====================
# 수정해 보세요!
# =====================
threshold = 120  # 기준값 변경 가능 (예: 100, 125, 140)
# =====================

# 기준값 초과 횟수 세기
count_exceed = 0
exceed_days  = []

for day_num, glucose in enumerate(blood_glucose_30days, start=1):
    if glucose > threshold:
        count_exceed += 1
        exceed_days.append((day_num, glucose))

print(f'혈당 기준값: {threshold} mg/dL')
print(f'기준값 초과 횟수: {count_exceed}회')
print(f'\n초과한 날짜와 수치:')
for day, val in exceed_days:
    print(f'  {day:2d}일차: {val} mg/dL')
```

```python
# 시각화
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'NanumGothic'
matplotlib.rcParams['axes.unicode_minus'] = False

fig, ax = plt.subplots(figsize=(12, 5))

days_x   = list(range(1, 31))
colors   = ['#e74c3c' if g > threshold else '#2ecc71' for g in blood_glucose_30days]

ax.bar(days_x, blood_glucose_30days, color=colors, edgecolor='white')
ax.axhline(y=threshold, color='#e74c3c', linestyle='--', linewidth=2, label=f'기준값 ({threshold} mg/dL)')
ax.set_title('30일치 공복혈당 데이터', fontsize=14)
ax.set_xlabel('측정일 (일차)')
ax.set_ylabel('혈당 (mg/dL)')
ax.legend()
ax.set_xticks(days_x)

plt.tight_layout()
plt.show()
print(f'빨간 막대: 기준값({threshold}) 초과 = {count_exceed}회')
```
