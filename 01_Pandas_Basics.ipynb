## 데이터란?

> 병원 환자 차트를 생각했을 때 환자 한 명의 이름·나이·혈압·체온이 하나의 행(row) 이고, 각 항목(이름·나이 등)이 열(column) 의미. Pandas는 거대한 차트를 컴퓨터에서 다루는 도구

### 의료 현장 데이터
- **전자의무기록(EMR):** 환자의 진료 기록, 처방, 검사 결과 등 수십 년치 데이터
- **웨어러블 기기:** 스마트워치가 실시간으로 측정하는 심박수·산소포화도 데이터
- **공공 보건 데이터:** 건강보험심사평가원의 전국 진료 통계 등

`Pandas`와 같은 도구를 활용해서 데이터를 분석할 수 있음

### 데이터 분석 파이프라인

```
데이터 수집(EMR/웨어러블) → 정제·표준화 → 탐색적 분석(EDA) → 모델링 → 의사결정 지원
```

## 가상의 환자 데이터 생성

실제 환자 데이터는 개인정보 보호법(데이터 3법)으로 인해 외부 공개가 매우 제한됨
오늘은 현실적인 수치를 가진 가상의 환자 데이터(Dummy Data) 를 직접 만들어 볼 것

````python
n_patients = 120  # 환자 수 (변경 가능)
random_seed = 42  # 랜덤 시드 (같은 숫자 = 같은 결과 보장)

np.random.seed(random_seed)

# 가상의 환자 데이터 생성
data = {
    '환자ID': [f'P{str(i).zfill(3)}' for i in range(1, n_patients + 1)],
    '나이': np.random.randint(18, 90, n_patients),
    '성별': np.random.choice(['남성', '여성'], n_patients),
    '체온_섭씨': np.round(np.random.normal(36.8, 0.5, n_patients), 1),
    '수축기혈압': np.random.randint(100, 180, n_patients),
    '이완기혈압': np.random.randint(60, 110, n_patients),
    '심박수': np.random.randint(55, 105, n_patients),
    '혈당_mg_dL': np.random.randint(70, 300, n_patients),
    '입원일수': np.random.randint(1, 30, n_patients),
    '진단명': np.random.choice(['고혈압', '당뇨', '폐렴', '골절', '심부전'], n_patients)
}

df = pd.DataFrame(data)
print(f'가상 환자 데이터 : 총 {len(df)}명')
````

## 데이터 크기: df.shape
- 행의 수, 열의 수

- **행(Row):** 환자 한 명 = 1개의 관찰값
- **열(Column):** 측정된 특성 (나이, 혈압 등) = 변수(Variable)

````python
# df.shape : (행의 수, 열의 수)를 튜플로 반환
print(f'데이터프레임 크기: {df.shape}')
print(f'총 환자 수 (행): {df.shape[0]}명')
print(f'총 변수 수 (열): {df.shape[1]}개')
````

## 데이터 필터링
- 특정 조건에 맞는 환자 찾기

````python
# 특정 열만 선택하기
print('--- 나이와 진단명만 보기 ---')
print(df[['나이', '진단명']].head())

# 조건으로 필터링 — 혈당이 200 이상인 환자만 추출
print('\n--- 혈당 200 이상 환자 ---')
high_glucose = df[df['혈당_mg_dL'] >= 200]
print(f'혈당 200 이상 환자 수: {len(high_glucose)}명')
high_glucose[['환자ID', '나이', '성별', '혈당_mg_dL', '진단명']].head()
````

````python
# 진단명별 환자 수 세기
print('--- 진단명별 환자 수 ---')
diagnosis_count = df['진단명'].value_counts()
print(diagnosis_count)
````
